```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import glob
from IPython.display import clear_output
import csv
import os
from PIL import Image 
from google.colab.patches import cv2_imshow
# 開啟 CSV 檔案
def show_img(img, bigger=False):
    if bigger:
        plt.figure(figsize=(15,15)) 
    image_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(image_rgb)
    plt.show()

def sharpen(img, sigma=150):    
    # sigma = 5、15、25
    blur_img = cv2.GaussianBlur(img, (0, 0), sigma)
    usm = cv2.addWeighted(img, 1.5, blur_img, -0.5, 0)
    
    return usm
def img_processing(img):
    # do something here
    img = sharpen(img)

    return img

with open('/content/drive/MyDrive/Traditional-Chinese-Scene-Character-Recognition-Contest-Advanced-Competition/YOLOv5/answer.csv', newline='') as csvfile:
     # 讀取 CSV 檔案內容
  rows = csv.reader(csvfile)
  z = 6000
  i=1
  # 以迴圈輸出每一列
  for row in rows:
    a = "/content/drive/MyDrive/Traditional-Chinese-Scene-Character-Recognition-Contest-Advanced-Competition/test/"
    #i=1
    img_path = a + row[0]+ '.jpg'
    print(img_path)
    img = cv2.imread(img_path)
    ana_txt_save_path = "/content/drive/MyDrive/char_to_string_test/"+str(row[0])
    if not os.path.exists(ana_txt_save_path):
        os.makedirs(ana_txt_save_path)
        i =1
    print(row[1:8])
    xl_up = float(row[1])#805
    yl_up = float(row[2])#617
                       
    xr_up = float(row[3])#1052
    yr_up = float(row[4])#613        
            
    xr_dn = float(row[5])#1059
    yr_dn = float(row[6])#662
            
    xl_dn = float(row[7])#809
    yl_dn = float(row[8])#665

    left = int(min(xl_up,xl_dn))
    right = int(max(xr_up,xr_dn))
    width = left+(right-left)

    up = int(min(yl_up,yr_up))
    down = int(max(yr_dn,yl_dn))
    height = up+(down-up)
    
    crop_img = img[up:height , left:width]
    #crop_img=img_processing(crop_img)
    #crop_img = cv2.resize(crop_img,interpolation=cv2.INTER_AREA)
    cv2_imshow(crop_img)
    #img_name ='./new1/'+str(i)
    if(i>=10):
      img_name =ana_txt_save_path+"/"+"00"+str(i)
    else:
      img_name =ana_txt_save_path+"/"+"000"+str(i)
    cv2.imwrite(img_name+'.jpg',crop_img)
    i = i+1
```
